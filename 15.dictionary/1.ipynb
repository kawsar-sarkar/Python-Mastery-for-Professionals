```python
# Creating a simple dictionary
student = {
    "name": "John",
    "age": 20,
    "grade": "A"
}

# 1. Basic value access using square brackets
print(student["name"])  # Output: John



```

```python
# 2. Using the get() method (safer way)
student = {
    "name": "John",
    "age": 20,
    "grade": "A"
}

print(student.get("age"))  # Output: 20
# get() with default value if key doesn't exist
print(student.get("phone", "Not found"))  # Output: Not found
```

```python
# 3. Accessing nested dictionary values
school_info = {
    "student": {
        "name": "John",
        "scores": {
            "math": 90,
            "science": 85
        }
    }
}

# Accessing nested values
print(school_info["student"]["name"])  # Output: John
print(school_info["student"]["scores"]["math"])  # Output: 90
```

```python
# 1. Dictionary with mixed data types
user = {
    "name": "Alice",
    "is_active": True,
    "hobbies": ["reading", "gaming", "swimming"],
    "address": {
        "street": "123 Main St",
        "city": "Boston"
    }
}

# Accessing list inside dictionary
print(user["hobbies"][0])  # Output: reading
print(user["address"]["city"])  # Output: Boston

# 2. Dictionary methods
car = {
    "brand": "Toyota",
    "model": "Camry",
    "year": 2020
}

# Get all keys
print(car.keys())  # Output: dict_keys(['brand', 'model', 'year'])

# Get all values
print(car.values())  # Output: dict_values(['Toyota', 'Camry', 2020])

# Get all key-value pairs
print(car.items())  # Output: dict_items([('brand', 'Toyota'), ('model', 'Camry'), ('year', 2020)])

# 3. Updating values
menu = {
    "coffee": 2.50,
    "tea": 2.00,
    "cake": 4.00
}

# Update single value
menu["coffee"] = 3.00

# Update multiple values at once
menu.update({"tea": 2.50, "cookie": 1.50})

# 4. Checking if key exists
inventory = {
    "apple": 50,
    "banana": 30,
    "orange": 40
}

# Using 'in' operator
if "apple" in inventory:
    print("Yes, apple is in stock")

# 5. Dictionary comprehension example
numbers = {x: x*x for x in range(5)}
print(numbers)  # Output: {0: 0, 1: 1, 2: 4, 3: 9, 4: 16}

# 6. Merging dictionaries (Python 3.5+)
dict1 = {"a": 1, "b": 2}
dict2 = {"c": 3, "d": 4}
merged = {**dict1, **dict2}
print(merged)  # Output: {'a': 1, 'b': 2, 'c': 3, 'd': 4}
```
